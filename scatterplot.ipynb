```python
import matplotlib.pyplot as plt

# Define the points
categories = ['Malware', 'ICS Compromise', 'Data Leak']
likelihood = [0.8, 0.8, 1]  # High likelihood
impact = [1, 0.8, 0.8]  # High impact

# Create the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(likelihood, impact, color='blue')
plt.title('Cybersecurity Threats: Likelihood vs. Impact')
plt.xlabel('Likelihood')
plt.ylabel('Impact')
plt.grid(True)

# Annotate the points
for i, category in enumerate(categories):
    plt.text(likelihood[i], impact[i], f" {category}", verticalalignment='bottom')

# Adjusting the limits to better visualize the high likelihood and high impact quadrant
plt.xlim(0, 1)
plt.ylim(0, 1)

# Highlight the high likelihood, high impact quadrant
plt.axhline(0.5, color='red', linestyle='--')
plt.axvline(0.5, color='red', linestyle='--')

plt.show()
```

```python
# Update the points with new categories
categories += ['Solar Flare', 'Domestic Terrorism', 'Extreme Flooding']
likelihood += [0.2, 0.2, 0.5]  # Low likelihood
impact += [1, 0.8, 0.5]  # High impact

# Recreate the scatter plot with updated data
plt.figure(figsize=(10, 6))
plt.scatter(likelihood[:3], impact[:3], color='red', label='High Likelihood, High Impact')
plt.scatter(likelihood[3:], impact[3:], color='orange', label='Low Likelihood, High Impact')
# plt.title('Cybersecurity Threats: Likelihood vs. Impact')
plt.xlabel('Likelihood')
plt.ylabel('Impact')
plt.grid(True)

# Annotate the points
for i, category in enumerate(categories):
    plt.text(likelihood[i], impact[i], f" {category}", verticalalignment='bottom')

# Adjusting the limits and highlight the quadrants
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.axhline(0.5, color='red', linestyle='--')
plt.axvline(0.5, color='red', linestyle='--')

plt.legend()
plt.show()
```
