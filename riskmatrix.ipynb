```python
import matplotlib.pyplot as plt

# Define the points
threats = [
    'Malware', 'ICS Compromise', 'Data Leak + Credential Stuffing', 'Extreme Flooding', 'Solar Flare', 
    'Domestic Terrorism', 'Heat Waves', 'Deep Freeze', 'Tornadoes', 'Collective Action'
]
likelihood = [0.8, 0.8, 1, 0.6, 0.2, 0.2, 1, 0.6, 0.8, 0.2]  # High likelihood
impact = [1, 0.8, 0.8, 0.6, 1, 0.8, 0.2, 0.4, 0.2, 0.2]  # High impact

# Create the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(likelihood[:4], impact[:4], color='red', label='High Likelihood, High Impact')
plt.scatter(likelihood[4:6], impact[4:6], color='orange', label='Low Likelihood, High Impact')
plt.scatter(likelihood[6:9], impact[6:9], color='yellow', label='High Likelihood, Low Impact')
plt.scatter(likelihood[9:], impact[9:], color='green', label='Low Likelihood, Low Impact')

plt.title('Texas Interconnection Risk Matrix\n')
plt.xlabel('Likelihood')
plt.ylabel('Impact')
plt.grid(False)

# Annotate the points
for i, category in enumerate(threats):
    plt.text(likelihood[i], impact[i], f" {category}", verticalalignment='bottom')

# Adjusting the limits to better visualize the high likelihood and high impact quadrant
plt.xlim(0, 1)
plt.ylim(0, 1)

plt.axhline(0.5, color='red', linestyle='--')
plt.axvline(0.5, color='red', linestyle='--')

plt.xticks([])  # Hide x-axis ticks
plt.yticks([])  # Hide y-axis ticks

plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.savefig("risk_matrix.png", bbox_inches='tight')
plt.show()
```
